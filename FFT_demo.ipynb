```python
# intialise - run this first

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import sounddevice as sd
```

Example: Chirp

```python
# A chirp signal is a signal in which the frequency increases (or decreases) over time.
# In this example, we generate a **linear chirp** where the frequency starts at 10 Hz and increases to 200 Hz over 1 second.

# Define parameters
fs = 1000  # Sampling frequency (Hz)
duration = 1  # Duration in seconds
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Generate chirp signal (frequency increases from 10 Hz to 200 Hz)
chirp_signal = signal.chirp(t, f0=10, f1=200, t1=duration, method='linear')

# A spectrogram is a time-frequency representation of a signal.
# It shows how the frequency content of a signal changes over time.
# To compute a spectrogram, we divide the signal into small overlapping time windows, apply the Fast Fourier Transform (FFT) to each segment, and plot the frequency content over time.

# Compute spectrogram using the Short-Time Fourier Transform (STFT)
frequencies, times, Sxx = signal.spectrogram(chirp_signal, fs)

# Plot the chirp signal waveform and its spectrogram
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# The first plot shows the chirp signal in the time domain. You can see how its frequency increases over time.
axes[0].plot(t, chirp_signal)
axes[0].set_title("Chirp Signal in Time Domain")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Amplitude")

# The second plot shows the spectrogram. The color intensity represents how much energy a frequency component has at a given time.
cax = axes[1].pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
fig.colorbar(cax, ax=axes[1], label='Power/Frequency (dB/Hz)')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("Frequency (Hz)")
axes[1].set_title("Spectrogram of the Chirp Signal")

plt.tight_layout()
plt.show()

# We can also save the chirp signal as an audio file so you can listen to it.
# The file is saved in WAV format and can be played with any media player.

# Normalize the signal for WAV format (convert to 16-bit integer)
chirp_signal_normalized = np.int16(chirp_signal / np.max(np.abs(chirp_signal)) * 32767)

# Save the chirp signal as a WAV file
wavfile.write("chirp_signal.wav", fs, chirp_signal_normalized)

# Play the generated chirp sound
sd.play(chirp_signal, samplerate=fs)
sd.wait()  # Wait until sound is finished playing

print("âœ… Chirp signal saved as 'chirp_signal.wav'")
```

Example: Aliased Chirp


ðŸš€ What is Aliasing?
	â€¢	When a signal is sampled below the Nyquist rate, high frequencies appear incorrectly as lower frequencies.
	â€¢	This happens because sampling folds frequencies back into the Nyquist range, causing distortion.

ðŸš€ What is an Aliased Chirp Signal?
	â€¢	A chirp signal increases in frequency over time.
	â€¢	If sampled too low, aliasing distorts the spectrogram by folding high frequencies into lower ones.

ðŸš€ How is the Spectrogram Computed?
	â€¢	The signal is divided into short time windows.
	â€¢	The Fast Fourier Transform (FFT) extracts frequency information from each window.
	â€¢	The spectrogram visualizes aliasing effects.

```python
# Define parameters
fs_low = 100  # ðŸš¨ Low Sampling Frequency (Aliasing Will Occur)
fs_high = 44100  # High Sampling Frequency for Playback
duration = 2  # Duration in seconds
f_chirp_start = 10  # Start Frequency in Hz
f_chirp_end = 500  # End Frequency in Hz (Above Nyquist Limit)

# Generate time vectors
t_low = np.linspace(0, duration, int(fs_low * duration), endpoint=False)
t_high = np.linspace(0, duration, int(fs_high * duration), endpoint=False)

# Generate the aliased chirp signal at low sampling rate
aliased_chirp_low = signal.chirp(t_low, f0=f_chirp_start, f1=f_chirp_end, t1=duration, method='linear')

# Resample the aliased chirp to a higher sampling rate for playback
aliased_chirp_high = np.interp(t_high, t_low, aliased_chirp_low)  # Interpolation

# Compute the spectrogram with fixed `nperseg`
frequencies, times, Sxx = signal.spectrogram(aliased_chirp_low, fs_low, nperseg=100)

# Plot waveform and spectrogram
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Time-domain representation of the aliased chirp
axes[0].plot(t_low, aliased_chirp_low)
axes[0].set_title("Aliased Chirp Signal in Time Domain")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Amplitude")

# Spectrogram representation showing aliasing
cax = axes[1].pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
fig.colorbar(cax, ax=axes[1], label='Power/Frequency (dB/Hz)')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("Frequency (Hz)")
axes[1].set_title("Spectrogram of an Aliased Chirp Signal")

plt.tight_layout()
plt.show()

# Normalize to int16 for WAV format
aliased_chirp_normalized = np.int16(aliased_chirp_high / np.max(np.abs(aliased_chirp_high)) * 32767)

# Save the aliased chirp as a WAV file
wavfile.write("aliased_chirp.wav", fs_high, aliased_chirp_normalized)

# Play the resampled aliased chirp
sd.play(aliased_chirp_high, samplerate=fs_high)
sd.wait()  # Wait until playback is finished

print("âœ… Aliased chirp saved as 'aliased_chirp.wav'")
```

Example: A music signal

```python
# ðŸ“¢ Now let's try to analyse the spectral components of a music signal.
# This script reads a WAV file (music track), computes its spectrogram, and visualizes the frequency content over time.

# Load the music file (Make sure you have a WAV file named 'music.wav' in the same directory)
fs, music_signal = wavfile.read("sax.wav")  # Load WAV file

# If the audio is stereo (2 channels), convert it to mono by averaging both channels
if len(music_signal.shape) > 1:
    music_signal = np.mean(music_signal, axis=1)  # Convert to mono

# Compute the spectrogram using the Short-Time Fourier Transform (STFT)
frequencies, times, Sxx = signal.spectrogram(music_signal, fs)

# Plot the waveform and spectrogram
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# The first plot shows the music waveform in the time domain.
time_axis = np.linspace(0, len(music_signal) / fs, num=len(music_signal))
axes[0].plot(time_axis, music_signal)
axes[0].set_title("Music Signal in Time Domain")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Amplitude")

# The second plot is the spectrogram, showing how the frequency content changes over time.
cax = axes[1].pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
fig.colorbar(cax, ax=axes[1], label='Power/Frequency (dB/Hz)')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("Frequency (Hz)")
axes[1].set_title("Spectrogram of the Music Track")

plt.tight_layout()
plt.show()

print("âœ… Spectrogram generated successfully!")

# Play the sound
sd.play(music_signal, samplerate=fs)
sd.wait()  # Wait until sound is finished playing
```

Example: Sine wave

```python
# Now let's try a sine wave

# Define parameters
fs = 1000  # Sampling frequency (Hz)
duration = 2  # Duration in seconds
f_sine = 50  # Frequency of the sine wave in Hz
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Generate the sine wave
sine_wave = np.sin(2 * np.pi * f_sine * t)

# Compute spectrogram
frequencies, times, Sxx = signal.spectrogram(sine_wave, fs)

# Plot waveform and spectrogram
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Time-domain representation of the sine wave
axes[0].plot(t, sine_wave)
axes[0].set_title("Sine Wave in Time Domain")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Amplitude")

# Spectrogram representation of the sine wave
cax = axes[1].pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
fig.colorbar(cax, ax=axes[1], label='Power/Frequency (dB/Hz)')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("Frequency (Hz)")
axes[1].set_title("Spectrogram of the Sine Wave")

plt.tight_layout()
plt.show()

# Normalize to int16 for WAV format
sine_wave_normalized = np.int16(sine_wave / np.max(np.abs(sine_wave)) * 32767)

# Save the sine wave as a WAV file
wavfile.write("sine_wave.wav", fs, sine_wave_normalized)


print("âœ… Sine wave saved as 'sine_wave.wav'")

# Play the sound
sd.play(sine_wave, samplerate=fs)
sd.wait()  # Wait until sound is finished playing
```
